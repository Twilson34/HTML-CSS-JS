### Mars Weather

```python
url="https://twitter.com/marswxreport?lang=en"
browser.visit(url)
```

```python
#Scraping image page
soup=bs(browser.html, 'html.parser')
```

```python
test=soup.find_all('div', class_="css-1dbjc4n")
test
```

```python
for t in test:
    try:
        div = t.find("div", class_='css-901oao r-jwli3a r-1qd0xha r-a023e6 r-16dba41 r-ad9z0x r-bcqeeo r-bnwqim r-qvutc0')
        if div == None:
            continue
        weather = div.find("span").get_text()
        if 'InSight' in weather:
             break
    except AttributeError:
        continue
```

```python
MarsWeather=weather.replace('\n', " ")
MarsWeather
```
